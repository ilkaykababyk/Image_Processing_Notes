```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
my_img = cv2.imread("./images.jpg",1)
my_img_gray = cv2.cvtColor(my_img, cv2.COLOR_BGR2GRAY)
```

```python
_,my_thresh_binary = cv2.threshold(my_img_gray, thresh =128, maxval=255, type=cv2.THRESH_BINARY)
```

```python
fig = plt.figure(figsize=(10,7))
rows=1
columns=2
fig.add_subplot(rows,columns,1)
plt.imshow(my_img_gray, cmap="gray")
plt.axis("off")
plt.title("Original Image")

fig.add_subplot(rows,columns,2)
plt.imshow(my_thresh_binary, cmap="gray")
plt.axis("off")
plt.title("BÄ°NARY THRESHOLDING Image")
```

```python
_,my_thresh_binary_inv = cv2.threshold(my_img_gray, thresh =128, maxval=255, type=cv2.THRESH_BINARY_INV)
```

```python
fig = plt.figure(figsize=(10,7))
rows=1
columns=3
fig.add_subplot(rows,columns,1)
plt.imshow(my_img_gray, cmap="gray")
plt.axis("off")
plt.title("Original Image")

fig.add_subplot(rows,columns,2)
plt.imshow(my_thresh_binary, cmap="gray")
plt.axis("off")
plt.title("BÄ°NARY THRESHOLDING Image")

fig.add_subplot(rows,columns,3)
plt.imshow(my_thresh_binary, cmap="gray")
plt.axis("off")
plt.title("BÄ°NARY THRESHOLDÄ°NG Image")
```

```python
_,my_thresh_trunc= cv2.threshold(my_img_gray, thresh =128, maxval=255, type=cv2.THRESH_TRUNC)
```

```python
fig = plt.figure(figsize=(10,7))
rows=1
columns=3
fig.add_subplot(rows,columns,1)
plt.imshow(my_img_gray, cmap="gray")
plt.axis("off")
plt.title("Original Image")

fig.add_subplot(rows,columns,2)
plt.imshow(my_thresh_binary, cmap="gray")
plt.axis("off")
plt.title("BÄ°NARY THRESHOLDING Image")

fig.add_subplot(rows,columns,3)
plt.imshow(my_thresh_binary_inv, cmap="gray")
plt.axis("off")
plt.title("BÄ°NARY THRESHOLDING Image")
```

```python
_,my_thresh_binary_trunc = cv2.threshold(my_img_gray, thresh =128, maxval=255, type=cv2.THRESH_TRUNC)
```

```python
fig = plt.figure(figsize=(10,7))
rows=2
columns=2
fig.add_subplot(rows,columns,1)
plt.imshow(my_img_gray, cmap="gray")
plt.axis("off")
plt.title("Original Image")

fig.add_subplot(rows,columns,2)
plt.imshow(my_thresh_binary, cmap="gray")
plt.axis("off")
plt.title("BÄ°NARY THRESHOLDING Image")

fig.add_subplot(rows,columns,3)
plt.imshow(my_thresh_binary, cmap="gray")
plt.axis("off")
plt.title("BÄ°NARY THRESHOLDING Image")

fig.add_subplot(rows,columns,4)
plt.imshow(my_thresh_binary, cmap="gray")
plt.axis("off")
plt.title("BÄ°NARY THRESHOLDING Image")

```

```python
_,my_thresh_tozero = cv2.threshold(my_img_gray, thresh =128, maxval=255, type=cv2.THRESH_TOZERO)
```
